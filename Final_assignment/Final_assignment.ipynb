Get the necessary data and store it in dataframes:

```python
# Locations of files to read:
Location_data = r'resources\ef2013a.csv'
    # Main data with gender and ethnicity information on students
     
Location_institution_data = r'resources\hd2013.csv'
    # File with institution data

#Grab the data from the main file with student numbers:
df_all_students = pd.read_csv(Location_data)

#Grab data from file with institution information:
df_institutions = pd.read_csv(Location_institution_data)
```

Now I will extract only the rows with an EFA level of 12 (which means it refers to graduate-level students).  After that, I will remove unnecessary columns from the dataframe in order to speed up future searches.<br/>
Unnecessary columns include all of the imputation values, because I've already searched through the .csv file for students, and I found that none of the imputation codes that appear disqualify the data, so they are irrelevant).  The columns with imputation values have names that start with the letter 'X' and then the name of the variable it is providing an imputation variable for.<br/>
To sum up: I'm dropping all the columns that I don't want.

```python
df_grad_students = df_all_students.loc[df_all_students['EFALEVEL']==12]

#next I'm going to remove some unnecessary columns to clean up the dataframe.
df_grad_students.drop('LINE',axis=1,inplace=True)
df_grad_students.drop('SECTION',axis=1,inplace=True)
df_grad_students.drop('LSTUDY',axis=1,inplace=True)
df_grad_students.drop('EFALEVEL',axis=1,inplace=True)

#now drop the imputation values (I already scanned through and determined that none of the ones used indicate an invalid value)
# the imputation values are in the columns that start with an 'X' (followed by the varname)
col_labels = df_grad_students.columns.values
for name in col_labels:
    if name[0]=='X':
        df_grad_students.drop(name,axis=1,inplace=True)
        
# I will also drop the columns dealing with students of unknown race, because they don't provide useful information:
df_grad_students.drop('EFUNKNT',axis=1,inplace=True)
df_grad_students.drop('EFUNKNM',axis=1,inplace=True)
df_grad_students.drop('EFUNKNW',axis=1,inplace=True)

print (df_grad_students[0:5])
```

Next is to strip down the dataframe with institution information so that it only contains the information I want (institution ID and FIPS state code).  I will also remove any institutions that are not in the 50 U.S. states

```python
#the relevant columns are: UNITID and FIPS (FIPS state code)
relevant_columns = ['UNITID','FIPS']
df_institutions = df_institutions[relevant_columns]

#Limit the set to only institutions in the 50 U.S. states
codes = pd.Series(FIPS_states.keys())
df_institutions = df_institutions[df_institutions.FIPS.isin(codes)]

print (df_institutions[0:5])  #Check the output
```

The step is combining the two dataframes (the one with the gender/ethnicity information for graduate students, and the one with the institution information).  I'm going to merge them (equivalent to a SQL <code>join</code> operation), using the UNITID from both dataframes as the keys. I'm doing an intersection-based merge so that the result will not include data for institutions that do not appear in the graduate student dataframe, and will also not include data from the graduate students whose institution ID does not appear in the institution dataframe.

```python
df_combined = df_grad_students.merge(df_institutions, 
                                     left_on='UNITID',
                                     right_on = 'UNITID', 
                                     how='inner')

#Drop the UNITID column, now that we've got the data linked to the states, it's no longer necessary.
df_combined.drop('UNITID',axis=1,inplace=True)


print (df_combined[0:5])
```

Now I will add another column for the state name (to replace the one with FIPS state code)

```python
df_combined["State"] = df_combined["FIPS"].map(FIPS_states)

#now drop the FIPS column, it is no longer necessary
df_combined.drop('FIPS',axis=1,inplace=True)

print (df_combined[0:5])
```

Next I need to merge the rows according to their FIPS state code (since I want state-wide data).  By using the .sum() function, I can get the sum of all institutions in a state.

```python
#Create a new df, grouped by State - also make State the index for the new df
df_totals_by_state = df_combined.groupby(['State']).sum()

print (df_totals_by_state[0:5])
```

Now I want to figure out what the national averages are (for % of graduate student body that is the target group, e.g. all women, or white men, or black women, etc).<br/>
To do this, I need to determine the ratios for each school.<br/>
Note: I need these averages in order to determine the threshold, and the range for the threshold.

```python
columns_to_copy = ['EFTOTLT','EFTOTLM','EFTOTLW', # Total
                   'EFAIANT','EFAIANM','EFAIANW', # American Indians or Alaskan Natives
                   'EFASIAT','EFASIAM','EFASIAW', # Asians
                   'EFBKAAT','EFBKAAM','EFBKAAW', # Black or African-Americans
                   'EFHISPT','EFHISPM','EFHISPW', # Hispanics
                   'EFNHPIT','EFNHPIM','EFNHPIW', # Native Hawaiian or other Pacific Islanders
                   'EFWHITT','EFWHITM','EFWHITW', # Whites
                   'EF2MORT','EF2MORM','EF2MORW', # 2 or more races
                   'EFNRALT','EFNRALM','EFNRALW'] # non-resident aliens

df_percentages_by_state = df_totals_by_state[columns_to_copy]
df_percentages_by_state = df_percentages_by_state.div(df_percentages_by_state.EFTOTLT,axis='index')
#Now the FIPS state code is the index

#Now drop the EFTOTLT column because it is no longer necessary:
df_percentages_by_state.drop('EFTOTLT',axis=1,inplace=True)

print (df_percentages_by_state.ix[0:5,0:2])
```

Now that I've created that dataframe with all the percentages for all institutions within each state, I will get the mean percentage and standard deviation for each category:

```python
national_avg_percentages = df_percentages_by_state.mean()
# national_avg_percentages.rename(EF_translation,inplace=True)
national_std_deviations = df_percentages_by_state.std()
# national_avg_percentages.rename(EF_translation,inplace=True)

total_percent_men = national_avg_percentages[0]
total_percent_women = national_avg_percentages[1]

print (national_avg_percentages)
```

##State visualizations

So, now that I've got some visualizations for the national data, I want to do some by state.  I will do a separate map for each category (e.g. one for total percent hispanics, one for percent hispanic men, etc).  What I'm going to do is have a U.S. map with the states marked out.  If a state's average is significantly less than the national average, it will be colored orange. The further away from the average, the more orange it will be. If a state's average is significantly more than the national average, it will be blue. The further from the average, the more blue it will be.  The difference from the average will be in terms of standard deviations - if the state average is within 1 standard deviation of the national average, it will be colored white.

The first step in creating these maps is to get the actual map background.  I'm just going to be doing the background for the lower 48 states because that's the most organized way to do it.<br/>
Now I'm going to make a new dataframe of state averages *without* the data for Alaska and Hawaii.

```python
#Make a new dataframe, so that we can work with the data without changing the original data
df_states = df_percentages_by_state

#Multiply by 100 to work with percentages instead of decimals
df_states = df_states * 100

#drop alaska and hawaii
df_states.drop('Alaska',axis=0,inplace=True)
df_states.drop('Hawaii',axis=0,inplace=True)


print (df_states[0:5]) #Check print out
```

**Note: My code for these maps is based on the fillstates.py example available in the matplotlib project on github, here:** https://github.com/matplotlib/basemap/tree/master/examples <br/>
My eternal gratitude goes to whoever coded that example!

I'm making a function (<code>determine_color</code>) to determine what the color of the state should be, based on how many standard deviations the state's percentage is away from the national percentage.<br/>
I'm also making a function (<code>make_map</code>) to create the national state-by-state map for a particular variable (based on the name of that variable, e.g. EFTOTLM for the total percent of men).

```python
#a series of reds, from lightest to darkest:
reds = ['#FFCCCC', '#FF9999','#FF6666','#FF3333','#FF0000','#CC0000','#990000','#660000']
blues = ['#CCFFFF','#9999FF','#6666FF','#3333FF','#0000FF','#0000CC','#000099','#000066']

#Determines the appropriate color of the state for the input column
def determine_color(stateName,colName):
    national_avg = national_avg_percentages[colName].item() * 100
    national_std = national_std_deviations[colName].item() * 100
    state_avg = df_states.loc[stateName,colName]
    maximum = df_states[colName].max()
    minimum = df_states[colName].min()
    
    #If within range of national average, color white:
    if (national_avg - national_std <= state_avg <= national_avg + national_std ):
        return '#FFFFFF'
    
    #Else if above the middle range, color red; the more standard deviations away the state average is from the national average,
    # the darker red the state should be.
    elif (state_avg > national_avg + national_std) :
        num_devs = int((state_avg  -national_avg)/national_std)
        if num_devs > len(reds):
            num_devs = len(reds) - 1
        return reds[num_devs]
    
    #Otherwise, state average is below the midle range, so color blue;
    # the more standard deviations away the state average is from the national average, the darker blue the state should be.
    else:
        num_devs = int((national_avg - state_avg)/national_std)
        if ( num_devs > len(blues)):
            num_devs = len(blues) - 1
        return blues[num_devs]
    
#Creates a map based on the data from the input column
def make_map(colName):
    plt.clf() # clear figure
    
    #Lambert Conformal map of lower 48 states
    m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95,suppress_ticks=True)

    #draw state boundaries (from U.S. census Bureau)
    shp_info = m.readshapefile('resources/st99_d00','states',drawbounds=True)
    
    statenames=[]

    for shapedict in m.states_info:
        statename = shapedict['NAME']
        # skip DC, Puerto Rico, Alaska, Hawaii
        if statename not in ['District of Columbia','Puerto Rico','Alaska','Hawaii']:
            stateval = df_states.loc[statename,colName]
        statenames.append(statename)
    
    #cycle through state names, color each one:
    ax = plt.gca()
    for nshape,seg in enumerate(m.states):
        if statenames[nshape] not in ['District of Columbia','Puerto Rico','Alaska','Hawaii']:
            color = determine_color(statenames[nshape],colName)
            poly = Polygon(seg,facecolor=color,edgecolor='black')
            ax.add_patch(poly)
    
    #create patches (keys) for the legend, and also create the legend:
    redpatch = mpatches.Patch(facecolor='red',edgecolor='black',label='Above national average')
    bluepatch = mpatches.Patch(facecolor='blue',edgecolor='black',label='Below national average')
    whitepatch = mpatches.Patch(facecolor='white',edgecolor='black',label='Within range of average')
    plt.legend(handles=[redpatch,bluepatch,whitepatch],loc=3,fontsize='small')
    
    #add a title:
    plt.title('Graduate students by state:\nPercent %s' % EF_translation[colName])
    
    #Add a subtitle:
    plt.suptitle('Note: The darker the color, the farther from average the represented percentage is.',y=0.02,horizontalalignment='center',verticalalignment='bottom')
    
    #Use tight_layout so that the map looks better
    plt.tight_layout()
```

Now I'm going to make a map for every column in the state dataframe.<br/>
Note: be patient, this <code>for</code> loop takes a while to finish because it has a lot of maps to make.

```python
# Create a map for each column of values, and save the maps.
for column in df_states:
    make_map(column)
    pylab.draw()
    pylab.savefig('resources/StateBy%s.png' % column,format='png')
```
